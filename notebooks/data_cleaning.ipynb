## Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/student_exam_scores.csv")
```

```python
# First, cap values above 50 (to handle the 51.3)
df.loc[df['exam_score'] > 50, 'exam_score'] = 50.0
# Scale the column by 2
df['exam_score'] = (df['exam_score'] * 2).round(1)
```

```python
df['hours_studied'] = df['hours_studied'].round(1)
df['sleep_hours'] = df['sleep_hours'].round(1)
df['attendance_percent'] = df['attendance_percent'].round(1)
```

```python
df['attendance_percent'] = df['attendance_percent'].clip(0, 100)
df['previous_scores'] = df['previous_scores'].clip(0, 100)
```

```python
# 7. Final check for duplicates and nulls (0 found in this set)
df = df.drop_duplicates()
df = df.dropna()
```

```python
from sklearn.preprocessing import StandardScaler
features = ['hours_studied', 'sleep_hours', 'attendance_percent', 'previous_scores']

# 3. Apply Z-score Standardization
# This transforms values to "number of standard deviations away from the mean"
scaler = StandardScaler()
df[features] = scaler.fit_transform(df[features]).round(4)
```

```python
df.to_csv('../data/cleaned_data.csv', index=False)
```

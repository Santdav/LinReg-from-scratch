```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/student_exam_scores.csv")

sns.set_theme(style="white", context="notebook")
plt.figure(figsize=(10, 4))
```

```python
print(df.head())
print(df.info())

df.columns
```

### Missing values or duplicates

```python
print(df.isnull().sum())
print(df.duplicated().sum())
```

## Univariable Analisys

### Study Hours

```python

f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)}, figsize=(10, 6))
 
# Add boxplot to the top row
sns.boxplot(x=df["hours_studied"], ax=ax_box, color="#69b3a2", width=0.5, fliersize=3)
ax_box.set(xlabel='') # Remove x-label from the boxplot so it doesn't overlap

# Add histogram + KDE to the bottom row
sns.histplot(data=df, x="hours_studied", kde=True, ax=ax_hist, color="#404040", bins=12)

# Clean up styling
sns.despine(ax=ax_hist)
sns.despine(ax=ax_box, left=True, bottom=True) # Hide the boxplot's borders
plt.tight_layout()
```
